### **Dropping missing values**

```python
# Treating Missing values (Based on team meeting/client discussion)
df_new = df.dropna()
df_new.isnull().sum()
```

```python
df_new.info()
```

```python
df_new.head()
```

```python
# converting cust_id float type to intger type
df_new['cust_id'] = df_new['cust_id'].astype('int64')
df_new.head()
```

```python
df_new.describe()
```

```python
df_new.describe().round(2)
```

## **removing negative values from quantity column**

```python
# conclusion : quantity column have some negative values
# we need to remove negative values
con = df_new.quantity > 0
con
```

```python
df_new = df_new[con]
df_new.info()
```

```python
df_new.describe().round(2)
```

```python
df_new.head()
```

## **Adding the column 'amount_spent'**

```python
df_new['amount_spent'] = df_new['quantity'] * df_new['unit_price']
df_new.head()
```

## **Rearranging columns for more readability**

```python
col_order = ['invoice_num','invoice_date','stock_code','description','quantity','unit_price','amount_spent','cust_id','country']
df_new = df_new[col_order]
df_new.head()
```

## **Accessing year and month value from invoice_date**


```python
year = df_new['invoice_date'].dt.year
month = df_new.invoice_date.dt.month
print(year,month)
df_new.insert(loc = 2, column = 'month', value = month) # adding month column
df_new
```

```python
# v1 = df_new['invoice_date'].map(lambda col: 100 * (col.year) + col.month)
v1 = 100 * (year) + month
df_new.insert(loc = 2, column = 'year_month', value = v1)
```

```python
df_new
```

## **Day of the week**
 In pandas, the day formate starts from 0 to 6

 Monday = 0, Tuesday = 1 .... Sunday = 6

 Apply +1 to make Monday = 1.....until Sunday = 7

```python
v2 = df_new.invoice_date.dt.dayofweek
v3 = v2 + 1
df_new.insert(loc = 4, column = 'day', value = v3)
df_new
v2.unique(),v2
```

```python
v3.unique(),v3
```

```python
df_new.head()
```

```python
df_new.insert(loc = 5, column = 'hour', value = df_new.invoice_date.dt.hour)
df_new.head()
```

# **Exploratory Data Analysis (EDA)**


## **Groupby**

```python
# groupby customer_id
df_new.groupby(by = ['cust_id']).count()
```

```python
# groupby using cust_id and country
df_new.groupby(by = ['cust_id', 'country']).count()
```

```python
df_new.groupby(by = ['cust_id', 'country'])['invoice_num'].count()
```

```python
# as_index = False
df_new.groupby(by =['cust_id','country'], as_index=False)['invoice_num'].count()
```

## **Data Visualization**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
orders = df_new.groupby (by=['cust_id','country'], as_index=False)['invoice_num'].count()
orders
```

### **number of orders for different customers**

```python
plt.subplots(figsize = (15,6))
plt.plot(orders.cust_id, orders.invoice_num)
plt.xlabel('Customer ID')
plt.ylabel('Number of Orders')
plt.title('Number of Orders for Different Customers')
plt.show()
```

### **check top 5 most no of orders**

```python
orders.sort_values(by='invoice_num', ascending=False).head()
```

### **Money spent by each customers**

```python
df_new.groupby(by=['cust_id','country'])['amount_spent'].sum()
```

```python
money_spent = df_new.groupby(by=['cust_id','country'], as_index=False)['amount_spent'].sum()
money_spent
```

**Top 5 customers who spent highest money**

```python
money_spent.sort_values(by='amount_spent',ascending=False).head()
```
