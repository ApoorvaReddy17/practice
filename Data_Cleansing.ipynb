# **data cleansing**

```python
# using regex removing spl charecters from Name column
emp['Name'] = emp['Name'].str.replace(r'\W', '', regex=True)
emp['Name']
```

```python
emp['Domain']=emp['Domain'].str.replace(r'\W', '', regex=True)
emp['Domain']
```

```python
emp['Age']=emp['Age'].str.extract(r'(\d+)')
emp
```

```python
emp['Salary']=emp['Salary'].str.replace(r'\W','',regex=True)
emp
```

```python
emp['Location']=emp['Location'].str.replace(r'\W','',regex=True)
emp
```

```python
emp['Exp']=emp['Exp'].str.extract(r'(\d+)')
emp
```

```python
clean_data=emp.copy()
```

```python
# clean_data.to_excel("clean_data.xlsx",index=False)
clean_data
```

# Handling Missing Values
# missing value treatment

```python
# this data contain num data and categorical data
# for num data we can replace with mean,median,mode strategy
# for categorical data we can replace with mode, KNN method
```

```python
clean_data.info()
```

```python
import numpy as np
# numpy for multidimension arrary
# pandas used for dataframes
```

```python
# missing value treatment for num value
clean_data['Age']=clean_data['Age'].astype(float)
clean_data['Exp']=clean_data['Exp'].astype(float)
clean_data['Salary']=clean_data['Salary'].astype(int)
clean_data.info()
```

```python
m=clean_data['Age'].mean() # m is mean of age
clean_data['Age']=clean_data['Age'].fillna(m)
```

```python
clean_data
```

```python
n=clean_data['Exp'].mean() # n is mean of Exp
clean_data['Exp']=clean_data['Exp'].fillna(n)
```

```python
clean_data
```

```python
# missing value treatment for cat data
x=clean_data['Location'].mode()
x
```

```python
clean_data['Location']=clean_data['Location'].fillna(x[0])
```

```python
clean_data.info()
```

```python
clean_data['Name']=clean_data['Name'].astype('category')
clean_data['Domain']=clean_data['Domain'].astype('category')
clean_data['Location']=clean_data['Location'].astype('category')
clean_data['Age']=clean_data['Age'].astype('int')
clean_data['Exp']=clean_data['Exp'].astype('int')

clean_data.info()
```

```python
# clean_data.to_excel("clean_data.xlsx",index=False)
# clean_data.to_csv("address_where_you_want_to_store\\clean_data.csv",index=False)
clean_data.to_csv("clean_data.csv",index=False)
clean_data
```
