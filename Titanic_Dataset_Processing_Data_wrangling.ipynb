```python
# creating Data Frame
import pandas as pd
df=pd.read_csv('titanic.csv')
df.head()
```

```python
# setting name column as index
df1=df.set_index(df['Name'])
df1.head()
```

```python
# after setting Name as index, accessing first row
df1.loc['Allen, Miss. Elisabeth Walton']
```

```python
# selecting rows based on condition
df1=df[df['Sex']=='female']
df1.head()
```

```python
# Selecting rows based on multiple conditions
df2=df[(df['Sex']=='female') & (df['Age']>=60)]
df2.head()
```

```python
# Replacing values
df3=df['Sex'].replace('female','Woman')
df3.head()
```

```python
# replacing values
df4=df['Sex'].replace(['female','male'],['Woman','Man'])
df4.head()
```

```python
df1=df.replace(1,"one")
df1.head()
```

```python
# renaming column names

df1= df.rename(columns={'Pclass':'Passenger Class','Sex':'Gender'})
df1.head()
```
