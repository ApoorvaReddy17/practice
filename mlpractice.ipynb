# **Feature Engineering**

```python
""" The Weekend and Revenue columns are currently set to Boolean values,
so we first need to convert into binary values. """

df['Weekend']=df['Weekend'].astype(int)
df['Revenue']=df['Revenue'].astype(int)
df.head()
```

```python
df['VisitorType'].unique()
```

```python
df['VisitorType'].value_counts()
```

```python
# returning_visitor or New_Visitor, For us one value is enough because other value is opposite to existing value.
# Lets add Returning_Visitor column to existing dataframe
df['Returning_Visitor'] = np.where(df['VisitorType']=='Returning_Visitor',1,0)
df.head()
```

```python
df['Returning_Visitor'].value_counts()
```

```python
# dropping VisitorType Column
# df.drop('VisitorType', axis = 1, inplace = True)
df = df.drop(columns = ['VisitorType'])
df.head()
```

```python
# checking Unique values in Month column
df['Month'].unique()
```

```python
# Applying Ordinal Encoding on Month column

oe = OrdinalEncoder()
df['Month'] = oe.fit_transform(df[['Month']])
df['Month'].unique()
```

```python
# Target variable is Revenue column, check value_counts() to get count of unique values
df['Revenue'].value_counts()
```
